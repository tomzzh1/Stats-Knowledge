```python
import numpy as np
import random
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
mean_out = []


for i in range(0,100):

    temp = np.random.uniform(-1,0,random.randint(15,40))
    
    mean_out.append(np.mean(temp))
```

```python
plt.title('Central Limit Theorem (Uniform Distribution)')
sns.histplot(mean_out)
plt.show()
```

```python
mean_out = []


for i in range(0,1000):

    temp = np.random.exponential(size = random.randint(20,60))
    
    mean_out.append(np.mean(temp))
```

```python
plt.title('Central Limit Theorem (Exponential Distribution)')
sns.histplot(mean_out)
plt.show()
```
